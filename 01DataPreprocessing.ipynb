```python
data.drop(columns = ['S. NO.', 'AUTHORS'], inplace = True) #removing the unrelavant columns
```

```python
#now we need to fix the source material column
data['SOURCE MATERIAL'].unique()
```

```python
def standardizeName(name):
    if type(name) == str:
        name_ = name.replace(' ', '').replace('-','').lower() 
        if 'and' in name: return 'mix'
        if name_ == 'classfflyash' : return 'Class F Fly-Ash'
        if name_ == 'classcflyash' : return 'Class C Fly-Ash'
        if name_ == 'slag' or name_ == 'ggbfs': return 'slag'
    return name #to check if it works
```

```python
data['SOURCE MATERIAL'] = data['SOURCE MATERIAL'].apply(standardizeName)
data['SOURCE MATERIAL'].unique()
```

```python
data['SOURCE MATERIAL'].fillna(method = "ffill" , inplace = True) #this method should fill all NaN with the preceeding valid values (for source material)
```

```python
data.dropna(inplace = True) #this should remove all records with null values
```
