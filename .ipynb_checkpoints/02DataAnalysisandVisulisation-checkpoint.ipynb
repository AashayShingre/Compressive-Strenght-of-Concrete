###### Now let's try to visualise the data, and try to understand some correlation, and distribution in data

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.distplot(data['FLY-ASH %'])
```

```python
#likewise, lets check univariate distribution for every column
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(12, 12))
fig.tight_layout()

for i, column in enumerate(data.columns):
    if i == 0 : continue #skip the first column
    j = i - 1
    sns.distplot(data[column],ax=axes[j//3,j%3], kde = False)
```

#### KDE plots (vs. compressive strength)

```python
import numpy as np
```

```python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 15))
fig.tight_layout()

cols = np.concatenate((data.columns[1:3], data.columns[4:]))
for i, column in enumerate(cols):
    sns.kdeplot(data[column], data['COMPRESSIVE STRENGTH (MPa)'], cmap="hot", shade=True, ax = axes[i//2, i%2])
```

```python
for i, column in enumerate(cols):
    sns.lmplot(column, 'COMPRESSIVE STRENGTH (MPa)', hue = "SOURCE MATERIAL", data = data)
```

#### Lets try some PCA (with two and 3 principal components)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(data.drop(['SOURCE MATERIAL', 'COMPRESSIVE STRENGTH (MPa)'], axis = 1))
```

Now we try pca

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_data[:5]
```

###### lets just build a scatter plot

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

###### lets try and plot it against compressive strength

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize = (15, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_data[:,0], pca_data[:,1], data['COMPRESSIVE STRENGTH (MPa)'], c='red', s=2)
# ax.view_init(30, 185)
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))
ax = fig.gca(projection='3d')
ax.plot_trisurf(pca_data[:,0], pca_data[:,1], data['COMPRESSIVE STRENGTH (MPa)'], cmap='viridis', linewidth=0.2)
plt.show()
```

```python
scaled_data_2 = scaler.fit_transform(data.drop(['SOURCE MATERIAL'], axis = 1))
pca_2 = PCA(n_components=3)
pca_data_2 = pca_2.fit_transform(scaled_data)
```

```python
fig = plt.figure(figsize = (15, 10))
ax = fig.gca(projection='3d')
ax.plot_trisurf(pca_data_2[:,0], pca_data_2[:,1], pca_data_2[:,2], cmap='viridis', linewidth=0.2)
ax.view_init(-10, 60)
plt.show()
```
